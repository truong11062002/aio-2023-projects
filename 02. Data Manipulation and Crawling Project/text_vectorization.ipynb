```python
import string

remove_characters = '\t”“' + string.punctuation
def text_normalize(text):
    text = text.lower()
    text = text.strip()
    text = text.replace('\n', ' ')
    for char in remove_characters:
        text = text.replace(char, '')
    
    return text
```

```python
text = " Nguyễn Nhật Trường"
print(text_normalize(text))
```

```python
def create_dictionary(corpus):
    dictionary = []
    for paragraph in corpus:
        paragraph = text_normalize(paragraph)
        tokens = paragraph.split()
        for token in tokens:
            if token not in dictionary:
                dictionary.append(token)

    return dictionary
```

```python
corpus = ["nguyen", "nguyen", "nhat", "trường"]
create_dictionary(corpus)
```

```python

def vectorize(text, dictionary, unknown_token_id):
    text = text_normalize(text)
    tokens = text.split()
    vector = [
        dictionary.index(token) \
        if token in dictionary else unknown_token_id \
            for token in tokens
    ]
    
    return vector
```

```python
dictionary = []
text = 'thư viện'
unknown_token_id = len(dictionary)
vectorize(text, dictionary, unknown_token_id) 
```

```python
import os

def read_files_txt(root_dir):
    corpus = []
    for filename in os.listdir(root_dir):
        filepath = os.path.join(
            root_dir,
            filename
        )

        with open(filepath, 'r') as f:
            text = f.read() 
            corpus.append(text)
    
    return corpus
```

```python
root_dir = '/content/vn_news_corpus'
corpus = read_files_txt(root_dir)
corpus
```

```python
dictionary = create_dictionary(corpus)
unknown_token_id = len(dictionary)
print(len(dictionary))
```
