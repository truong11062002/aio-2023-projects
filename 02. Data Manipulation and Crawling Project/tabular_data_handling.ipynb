```python
tv_lst = [lst[0] for lst in advertising_lst]
radio_lst = [lst[1] for lst in advertising_lst]
newspaper_lst = [lst[2] for lst in advertising_lst]
sales_lst = [lst[3] for lst in advertising_lst]

tv_avg = sum(tv_lst) / len(tv_lst)
radio_avg = sum(radio_lst) / len(radio_lst)
newspaper_avg = sum(newspaper_lst) / len(newspaper_lst)
sales_avg = sum(sales_lst) / len(sales_lst)

print(f'TV Average: {tv_avg}')
print(f'Radio Average: {radio_avg}')
print(f'Newspaper Average: {newspaper_avg}')
print(f'Sales Average: {sales_avg}')
```

```python
def median(lst):
    sorted_lst = sorted(lst)

    n = len(lst)
    mid = n // 2
    if n % 2 == 0:
        y = (sorted_lst[mid] + sorted_lst[mid - 1]) / 2
    else:
        y = sorted_lst[mid]

    return y

tv_lst = [lst[0] for lst in advertising_lst]
radio_lst = [lst[1] for lst in advertising_lst]
newspaper_lst = [lst[2] for lst in advertising_lst]
sales_lst = [lst[3] for lst in advertising_lst]

tv_median = median(tv_lst)
radio_median = median(radio_lst)
newspaper_median = median(newspaper_lst)
sales_median = median(sales_lst)

print(f'TV Median: {tv_median}')
print(f'Radio Median: {radio_median}')
print(f'Newspaper Median: {newspaper_median}')
print(f'Sales Median: {sales_median}')
```
